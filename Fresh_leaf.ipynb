```python
import glob
import cv2 as cv
import pandas as pd
import numpy as np
images4=[cv.imread(file,cv.IMREAD_UNCHANGED) for file in sorted(glob.glob(r"/content/drive/MyDrive/ML_DATASET/papaya/Fresh Papaya Leaf/clahe/*.png"))]
```

```python
! pip install -q tensorflow-model-optimization
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import tensorflow_model_optimization as tfmot
li=[]

for i,img in enumerate(images4):
    model = Sequential()
    model.add(Dense(1024, activation='relu', kernel_initializer='he_normal', input_shape=(65536,)))
    
    model.add(Dense(512, activation='relu', kernel_initializer='he_normal'))
    model1 =tfmot.sparsity.keras.prune_low_magnitude(model)
    # compile the model
    model1.compile()
    # fit the model
    img=img.reshape(1,256*256)
    img=img.astype('float64')

  
    from tensorflow.keras import backend as K
    layerIndex = 1
    func = K.function([model1.get_layer(index=0).input], model1.get_layer(index=layerIndex).output)
    layerOutput = func([img])  # input_data is a numpy array
    
    li.append(layerOutput[0])
```

```python
df=pd.DataFrame(li)
```

```python
col=df.columns
cnt =0
for i in col:
  if(df[i]==0):
    cnt=cnt+1

print(cnt)
```
