```python
import glob
import cv2 as cv
import pandas as pd
import numpy as np
images4=[cv.imread(file,cv.IMREAD_UNCHANGED) for file in sorted(glob.glob(r"/content/drive/MyDrive/ML_DATASET/papaya/Papaya Leaf Curl/clahe/*.png"))]
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
li=[]

for i,img in enumerate(images4):
    model = Sequential()
    model.add(Dense(1024, activation='relu', kernel_initializer='he_normal', input_shape=(65536,)))
    model.add(Dense(512, activation='relu', kernel_initializer='he_normal'))
    # compile the model
    model.compile()
    # fit the model
    img=img.reshape(1,256*256)
    img=img.astype('float64')
  
    from tensorflow.keras import backend as K
    layerIndex = 1
    func = K.function([model.get_layer(index=0).input], model.get_layer(index=layerIndex).output)
    layerOutput = func([img])  # input_data is a numpy array
    
    li.append(layerOutput[0])
```

```python
df=pd.DataFrame(li)
```
